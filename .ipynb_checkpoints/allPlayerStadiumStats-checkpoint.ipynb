```python
def ByInnings(df,current_innings):
    
    df = df[df.innings == current_innings]
    df.reset_index(inplace = True,drop = True)
    
    df['isDot'] = df['runs_of_bat'].apply(lambda x: 1 if x == 0 else 0)
    df['isOne'] = df['runs_of_bat'].apply(lambda x: 1 if x == 1 else 0)
    df['isTwo'] = df['runs_of_bat'].apply(lambda x: 1 if x == 2 else 0)
    df['isThree'] = df['runs_of_bat'].apply(lambda x: 1 if x == 3 else 0)
    df['isFour'] = df['runs_of_bat'].apply(lambda x: 1 if x == 4 else 0)
    df['isSix'] = df['runs_of_bat'].apply(lambda x: 1 if x == 6 else 0)
    
    runs = pd.DataFrame(df.groupby(['batsman'])['runs_of_bat'].sum()).reset_index().rename(columns={'runs_of_bat':'runs'})
    innings = pd.DataFrame(df.groupby(['batsman'])['match_id'].apply(lambda x: len(list(np.unique(x))))).reset_index().rename(columns={'match_id':'innings'})
    balls = pd.DataFrame(df.groupby(['batsman'])['match_id'].count()).reset_index().rename(columns={'match_id':'balls'})
    dismissals = pd.DataFrame(df.groupby(['batsman'])['player_dismissed'].count()).reset_index().rename(columns={'player_dismissed':'dismissals'})
    sixes = pd.DataFrame(df.groupby(['batsman'])['isSix'].sum()).reset_index().rename(columns = {'isSix':'sixes'})
    fours = pd.DataFrame(df.groupby(['batsman'])['isFour'].sum()).reset_index().rename(columns = {'isFour':'fours'})
    
    df = pd.merge(innings,runs,on = 'batsman').merge( balls, on = 'batsman').merge( dismissals, on = 'batsman').merge( fours, on = 'batsman').merge( sixes, on = 'batsman')
    df['RPI'] =df.apply(lambda x: (x['runs']/x['innings']),axis = 1)
    return df
```

```python
def balls_per_dismissal(balls, dismissals):
    if dismissals > 0:
        return balls/dismissals
    else:
        return balls/1 
    
def balls_per_boundary(balls, boundaries):
    if boundaries > 0:
        return balls/boundaries
    else:
        return balls/1 
```

```python
def ByCustom(df, current_venue, current_phase, current_opposition):
    
    df = df[df.venue == current_venue]
    df = df[df.phase == current_phase]
    df = df[df.bowling_team == current_opposition]
   
  
    df.reset_index(inplace = True,drop = True)
    
    df['isDot'] = df['runs_of_bat'].apply(lambda x: 1 if x == 0 else 0)
    df['isOne'] = df['runs_of_bat'].apply(lambda x: 1 if x == 1 else 0)
    df['isTwo'] = df['runs_of_bat'].apply(lambda x: 1 if x == 2 else 0)
    df['isThree'] = df['runs_of_bat'].apply(lambda x: 1 if x == 3 else 0)
    df['isFour'] = df['runs_of_bat'].apply(lambda x: 1 if x == 4 else 0)
    df['isSix'] = df['runs_of_bat'].apply(lambda x: 1 if x == 6 else 0)
    
    runs = pd.DataFrame(df.groupby(['batsman'])['runs_of_bat'].sum()).reset_index().rename(columns={'runs_of_bat':'runs'})
    innings = pd.DataFrame(df.groupby(['batsman'])['match_id'].apply(lambda x: len(list(np.unique(x))))).reset_index().rename(columns={'match_id':'innings'})
    balls = pd.DataFrame(df.groupby(['batsman'])['match_id'].count()).reset_index().rename(columns={'match_id':'balls'})
    dismissals = pd.DataFrame(df.groupby(['batsman'])['player_dismissed'].count()).reset_index().rename(columns={'player_dismissed':'dismissals'})
    sixes = pd.DataFrame(df.groupby(['batsman'])['isSix'].sum()).reset_index().rename(columns = {'isSix':'sixes'})
    fours = pd.DataFrame(df.groupby(['batsman'])['isFour'].sum()).reset_index().rename(columns = {'isFour':'fours'})

    dots = pd.DataFrame(df.groupby(['batsman'])['isDot'].sum()).reset_index().rename(columns = {'isDot':'dots'})

    
    df = pd.merge(innings,runs,on = 'batsman').merge( balls, on = 'batsman').merge( dismissals, on = 'batsman').merge( fours, on = 'batsman').merge( sixes, on = 'batsman').merge( dots, on = 'batsman')
    df['RPI'] = df.apply(lambda x: x['runs']/x['innings'], axis = 1)
    df['SR'] = df.apply(lambda x: 100*(x['runs']/x['balls']),axis = 1)
    df['BPD'] = df.apply(lambda x: balls_per_dismissal(x['balls'], x['dismissals']), axis = 1)
    df['BPB'] = df.apply(lambda x: balls_per_boundary(x['balls'], (x['fours'] + x['sixes'])), axis = 1)

    df['dot_percentage'] = df.apply(lambda x: x['dots']/x['balls'], axis = 1)

    return df
```

```python
def get_phase(over_no):
    if over_no < 6:
        return'PowerPlay'
        
    elif over_no < 15:
        return'Middle Over'
    else:
        return'Death Over'
```

```python
from fuzzywuzzy import process
import pandas as pd


# List of batsmen names you want to filter
batsmen_list = ['YBK Jaiswal', 'V Kohli', 'F du Plessis']

# Define a function for fuzzy filtering
def fuzzy_filter(df, column_name, names_list):
    # Get unique values from the column
    unique_values = df[column_name].unique()

    # Find the best matches for each name in the list
    matched_names = []
    for name in names_list:
        match = process.extractOne(name, unique_values)
        if match and match[1] > 80:  # Threshold for fuzzy matching
            matched_names.append(match[0])

    # Filter the DataFrame for the matched names
    return df[df[column_name].isin(matched_names)]

# Apply fuzzy filtering and sort by runs
filtered_df = fuzzy_filter(df, 'batsman', batsmen_list).sort_values(by='runs', ascending=False)

# Display the filtered DataFrame
print(filtered_df)
```

```python
def ByVenueAndOpposition(df, current_venue, current_opposition):
    
    # Filter by venue and opposition only (removed phase filter)
    df = df[df.venue == current_venue]
    df = df[df.bowling_team == current_opposition]
   
    df.reset_index(inplace = True, drop = True)
    
    # Create binary indicators for different run outcomes
    df['isDot'] = df['runs_of_bat'].apply(lambda x: 1 if x == 0 else 0)
    df['isOne'] = df['runs_of_bat'].apply(lambda x: 1 if x == 1 else 0)
    df['isTwo'] = df['runs_of_bat'].apply(lambda x: 1 if x == 2 else 0)
    df['isThree'] = df['runs_of_bat'].apply(lambda x: 1 if x == 3 else 0)
    df['isFour'] = df['runs_of_bat'].apply(lambda x: 1 if x == 4 else 0)
    df['isSix'] = df['runs_of_bat'].apply(lambda x: 1 if x == 6 else 0)
    
    # Aggregate statistics by batsman
    runs = pd.DataFrame(df.groupby(['batsman'])['runs_of_bat'].sum()).reset_index().rename(columns={'runs_of_bat':'runs'})
    innings = pd.DataFrame(df.groupby(['batsman'])['match_id'].apply(lambda x: len(list(np.unique(x))))).reset_index().rename(columns={'match_id':'innings'})
    balls = pd.DataFrame(df.groupby(['batsman'])['match_id'].count()).reset_index().rename(columns={'match_id':'balls'})
    dismissals = pd.DataFrame(df.groupby(['batsman'])['player_dismissed'].count()).reset_index().rename(columns={'player_dismissed':'dismissals'})
    sixes = pd.DataFrame(df.groupby(['batsman'])['isSix'].sum()).reset_index().rename(columns = {'isSix':'sixes'})
    fours = pd.DataFrame(df.groupby(['batsman'])['isFour'].sum()).reset_index().rename(columns = {'isFour':'fours'})
    dots = pd.DataFrame(df.groupby(['batsman'])['isDot'].sum()).reset_index().rename(columns = {'isDot':'dots'})
    
    # Merge all statistics into a single DataFrame
    df = pd.merge(innings,runs,on = 'batsman').merge(balls, on = 'batsman').merge(dismissals, on = 'batsman').merge(fours, on = 'batsman').merge(sixes, on = 'batsman').merge(dots, on = 'batsman')
    
    # Calculate performance metrics
    df['RPI'] = df.apply(lambda x: x['runs']/x['innings'], axis = 1)
    df['SR'] = df.apply(lambda x: 100*(x['runs']/x['balls']),axis = 1)
    df['BPD'] = df.apply(lambda x: balls_per_dismissal(x['balls'], x['dismissals']), axis = 1)
    df['BPB'] = df.apply(lambda x: balls_per_boundary(x['balls'], (x['fours'] + x['sixes'])), axis = 1)
    df['dot_percentage'] = df.apply(lambda x: x['dots']/x['balls'], axis = 1)

    return df
```

```python
from fuzzywuzzy import process

def ByVenueOnly(df, venue_names, threshold=85):
    """
    Modified to accept multiple venue names with fuzzy matching
    """
    # Get unique venue names from DataFrame
    unique_venues = df['venue'].unique()
    
    # Find matching venues using fuzzy logic
    matched_venues = []
    unmatched_venues = []
    
    if isinstance(venue_names, str):
        venue_names = [venue_names]  # Convert single string to list
        
    for venue in venue_names:
        match = process.extractOne(venue, unique_venues)
        if match and match[1] >= threshold:
            matched_venues.append(match[0])
        else:
            unmatched_venues.append(venue)
    
    # Print warnings for unmatched venues
    if unmatched_venues:
        print(f"No matches found for venues: {', '.join(unmatched_venues)} (threshold: {threshold})")
    
    # Filter DataFrame by matched venues
    df = df[df['venue'].isin(matched_venues)]
    
    # Rest of your original processing
    df.reset_index(inplace=True, drop=True)
    
    # Create binary indicators for different run outcomes
    df['isDot'] = df['runs_of_bat'].apply(lambda x: 1 if x == 0 else 0)
    df['isOne'] = df['runs_of_bat'].apply(lambda x: 1 if x == 1 else 0)
    df['isTwo'] = df['runs_of_bat'].apply(lambda x: 1 if x == 2 else 0)
    df['isThree'] = df['runs_of_bat'].apply(lambda x: 1 if x == 3 else 0)
    df['isFour'] = df['runs_of_bat'].apply(lambda x: 1 if x == 4 else 0)
    df['isSix'] = df['runs_of_bat'].apply(lambda x: 1 if x == 6 else 0)
    
    # Aggregate statistics by batsman
    runs = pd.DataFrame(df.groupby(['batsman'])['runs_of_bat'].sum()).reset_index().rename(columns={'runs_of_bat': 'runs'})
    innings = pd.DataFrame(df.groupby(['batsman'])['match_id'].apply(lambda x: len(list(np.unique(x))))).reset_index().rename(columns={'match_id': 'innings'})
    balls = pd.DataFrame(df.groupby(['batsman'])['match_id'].count()).reset_index().rename(columns={'match_id': 'balls'})
    dismissals = pd.DataFrame(df.groupby(['batsman'])['player_dismissed'].count()).reset_index().rename(columns={'player_dismissed': 'dismissals'})
    sixes = pd.DataFrame(df.groupby(['batsman'])['isSix'].sum()).reset_index().rename(columns={'isSix': 'sixes'})
    fours = pd.DataFrame(df.groupby(['batsman'])['isFour'].sum()).reset_index().rename(columns={'isFour': 'fours'})
    dots = pd.DataFrame(df.groupby(['batsman'])['isDot'].sum()).reset_index().rename(columns={'isDot': 'dots'})
    
    # Merge all statistics into a single DataFrame
    df = pd.merge(innings, runs, on='batsman').merge(balls, on='batsman').merge(dismissals, on='batsman').merge(fours, on='batsman').merge(sixes, on='batsman').merge(dots, on='batsman')
    
    # Calculate performance metrics
    df['RPI'] = df.apply(lambda x: x['runs'] / x['innings'], axis=1)
    df['SR'] = df.apply(lambda x: 100 * (x['runs'] / x['balls']), axis=1)
    df['BPD'] = df.apply(lambda x: balls_per_dismissal(x['balls'], x['dismissals']), axis=1)
    df['BPB'] = df.apply(lambda x: balls_per_boundary(x['balls'], (x['fours'] + x['sixes'])), axis=1)
    df['dot_percentage'] = df.apply(lambda x: x['dots'] / x['balls'], axis=1)

    return df
```

```python
from fuzzywuzzy import process
import pandas as pd

# Define a function for fuzzy filtering
def fuzzy_filter(df, column_name, names_list, threshold=70):
    # Get unique values from the column
    unique_values = df[column_name].unique()

    # Find the best matches for each name in the list
    matched_names = []
    unmatched_names = []  # To track names that don't match

    for name in names_list:
        match = process.extractOne(name, unique_values)
        if match and match[1] >= threshold:  # Threshold for fuzzy matching
            matched_names.append(match[0])  # Append only the matched name
        else:
            unmatched_names.append(name)  # Append unmatched name

    # Print unmatched batsmen names
    if unmatched_names:
        print(f"No matching data found for the following batsmen: {', '.join(unmatched_names)}")

    # Filter the DataFrame for the matched names
    return df[df[column_name].isin(matched_names)]

# Example usage
# Assuming 'df' is your DataFrame and has a column named 'batsman'
batsmen_list = [
   'Faf du Plessis', 'Jake Fraser-McGurk', 'Abishek Porel', 
    'Karun Nair', 'KL Rahul', 'Tristan Stubbs', 
    'Ashutosh Sharma', 'Axar Patel', 'Vipraj Nigam', 
    'Mitchell Starc','S Samson', 'Y Jaiswal', 'Nitish Rana', 
    'Riyan Parag', 'D Jurel', 'SM Hetmyer', 
    'Wanindu Hasaranga', 'J Archer', 'Maheesh Theekshana'
]

# Apply fuzzy filtering and sort by runs
filtered_df = fuzzy_filter(df, 'batsman', batsmen_list).sort_values(by='RPI', ascending=False)

# Display the filtered DataFrame
#print(filtered_df)
```
