## Problem Statement:

### Auction: Players who can fulfil the role of pinch hitter in powerplay for SRH?

- Player analysis
- Team analysis
- Batsmen evaluation (ranking) for that particular role

### Keywords: Pinch hitter, Powerplay, SRH
- Quantifying these words:
    - Pinch Hitter = High SR, High balls_per_boundary, Decent Runs, Decent Balls Faced
    - Powerplay = 1-6 overs
    - SRH = Analysis By Team

## Terminology Used:
- SR = Strike Rate = Scoring rate for each ball
- RPI = Runs Per Innings = Runs scored per each innings including outs/not-outs
- BPB = Balls Per Boundary = No of balls for each six or a four
- BPD = Balls Per Dismissal = Average number of balls faced per each dismissal

## Part 1: Player Statistics

### 1.1 : Extract player-level statistics
### 1.2 : Determine player type from the above stats

```python
def balls_per_dismissal(balls, dismissals):
    if dismissals > 0:
        return balls/dismissals
    else:
        return balls/1 
    
def balls_per_boundary(balls, boundaries):
    if boundaries > 0:
        return balls/boundaries
    else:
        return balls/1 

def playerStatistics(df):
    df.reset_index(inplace = True, drop = True)
    
    df['isDot'] = df['batsman_runs'].apply(lambda x: 1 if x == 0 else 0)
    df['isOne'] = df['batsman_runs'].apply(lambda x: 1 if x == 1 else 0)
    df['isTwo'] = df['batsman_runs'].apply(lambda x: 1 if x == 2 else 0)
    df['isThree'] = df['batsman_runs'].apply(lambda x: 1 if x == 3 else 0)
    df['isFour'] = df['batsman_runs'].apply(lambda x: 1 if x == 4 else 0)
    df['isSix'] = df['batsman_runs'].apply(lambda x: 1 if x == 6 else 0)
    
    runs = pd.DataFrame(df.groupby(['batsman', 'match_id'])['batsman_runs'].sum().reset_index()).groupby(['batsman'])['batsman_runs'].sum().reset_index().rename(columns={'batsman_runs':'runs'})
    innings = pd.DataFrame(df.groupby(['batsman'])['match_id'].apply(lambda x: len(list(np.unique(x)))).reset_index()).rename(columns = {'match_id':'innings'})
    balls = pd.DataFrame(df.groupby(['batsman'])['match_id'].count()).reset_index().rename(columns = {'match_id':'balls'})
    dismissals = pd.DataFrame(df.groupby(['batsman'])['player_dismissed'].count()).reset_index().rename(columns = {'player_dismissed':'dismissals'})
    
    dots = pd.DataFrame(df.groupby(['batsman'])['isDot'].sum()).reset_index().rename(columns = {'isDot':'dots'})
    ones = pd.DataFrame(df.groupby(['batsman'])['isOne'].sum()).reset_index().rename(columns = {'isOne':'ones'})
    twos = pd.DataFrame(df.groupby(['batsman'])['isTwo'].sum()).reset_index().rename(columns = {'isTwo':'twos'})
    threes = pd.DataFrame(df.groupby(['batsman'])['isThree'].sum()).reset_index().rename(columns = {'isThree':'threes'})
    fours = pd.DataFrame(df.groupby(['batsman'])['isFour'].sum()).reset_index().rename(columns = {'isFour':'fours'})
    sixes = pd.DataFrame(df.groupby(['batsman'])['isSix'].sum()).reset_index().rename(columns = {'isSix':'sixes'})
    
    df = pd.merge(innings, runs, on = 'batsman').merge(balls, on = 'batsman').merge(dismissals, on = 'batsman').merge(dots, on = 'batsman').merge(ones, on = 'batsman').merge(twos, on = 'batsman').merge(threes, on = 'batsman').merge(fours, on = 'batsman').merge(sixes, on = 'batsman')
    
    #StrikeRate
    df['SR'] = df.apply(lambda x: 100*(x['runs']/x['balls']), axis = 1)

    #runs per innings
    df['RPI'] = df.apply(lambda x: x['runs']/x['innings'], axis = 1)

    #balls per dismissals
    df['BPD'] = df.apply(lambda x: balls_per_dismissal(x['balls'], x['dismissals']), axis = 1)

    #balls per boundary
    df['BPB'] = df.apply(lambda x: balls_per_boundary(x['balls'], (x['fours'] + x['sixes'])), axis = 1)
    
    return df
```

```python
df = playerStatistics(deliveries)
```

### 1.3 : Performance in different phases of play

```python
def phase(over):
    if over <= 6:
        return 'Powerplay'
    elif over <= 15:
        return 'Middle'
    else:
        return 'Death'
```

```python
deliveries['phase'] = deliveries['over'].apply(lambda x: phase(x))
```

```python
def phasesOfplay(df, current_phase):
    
    df = df[df.phase == current_phase]
    df.reset_index(inplace = True, drop = True)
    
    df['isDot'] = df['batsman_runs'].apply(lambda x: 1 if x == 0 else 0)
    df['isOne'] = df['batsman_runs'].apply(lambda x: 1 if x == 1 else 0)
    df['isTwo'] = df['batsman_runs'].apply(lambda x: 1 if x == 2 else 0)
    df['isThree'] = df['batsman_runs'].apply(lambda x: 1 if x == 3 else 0)
    df['isFour'] = df['batsman_runs'].apply(lambda x: 1 if x == 4 else 0)
    df['isSix'] = df['batsman_runs'].apply(lambda x: 1 if x == 6 else 0)
    
    runs = pd.DataFrame(df.groupby(['batsman', 'match_id'])['batsman_runs'].sum().reset_index()).groupby(['batsman'])['batsman_runs'].sum().reset_index().rename(columns={'batsman_runs':'runs'})
    innings = pd.DataFrame(df.groupby(['batsman'])['match_id'].apply(lambda x: len(list(np.unique(x)))).reset_index()).rename(columns = {'match_id':'innings'})
    balls = pd.DataFrame(df.groupby(['batsman'])['match_id'].count()).reset_index().rename(columns = {'match_id':'balls'})
    dismissals = pd.DataFrame(df.groupby(['batsman'])['player_dismissed'].count()).reset_index().rename(columns = {'player_dismissed':'dismissals'})
    
    dots = pd.DataFrame(df.groupby(['batsman'])['isDot'].sum()).reset_index().rename(columns = {'isDot':'dots'})
    ones = pd.DataFrame(df.groupby(['batsman'])['isOne'].sum()).reset_index().rename(columns = {'isOne':'ones'})
    twos = pd.DataFrame(df.groupby(['batsman'])['isTwo'].sum()).reset_index().rename(columns = {'isTwo':'twos'})
    threes = pd.DataFrame(df.groupby(['batsman'])['isThree'].sum()).reset_index().rename(columns = {'isThree':'threes'})
    fours = pd.DataFrame(df.groupby(['batsman'])['isFour'].sum()).reset_index().rename(columns = {'isFour':'fours'})
    sixes = pd.DataFrame(df.groupby(['batsman'])['isSix'].sum()).reset_index().rename(columns = {'isSix':'sixes'})
    
    df = pd.merge(innings, runs, on = 'batsman').merge(balls, on = 'batsman').merge(dismissals, on = 'batsman').merge(dots, on = 'batsman').merge(ones, on = 'batsman').merge(twos, on = 'batsman').merge(threes, on = 'batsman').merge(fours, on = 'batsman').merge(sixes, on = 'batsman')
    
    #StrikeRate
    df['SR'] = df.apply(lambda x: 100*(x['runs']/x['balls']), axis = 1)

    #runs per innings
    df['RPI'] = df.apply(lambda x: x['runs']/x['innings'], axis = 1)

    #balls per dismissals
    df['BPD'] = df.apply(lambda x: balls_per_dismissal(x['balls'], x['dismissals']), axis = 1)

    #balls per boundary
    df['BPB'] = df.apply(lambda x: balls_per_boundary(x['balls'], (x['fours'] + x['sixes'])), axis = 1)
    
    return df
```

```python
pp_df = phasesOfplay(deliveries, 'Powerplay')
mid_df = phasesOfplay(deliveries, 'Middle')
dth_df = phasesOfplay(deliveries, 'Death')
```

### Pair wise comparision matrices using Saaty scale & then arriving at weights for each parameter

```python
wt_sr, wt_rpi, wt_bpd, wt_bpb = 0.38, 0.25, 0.12, 0.26
```

```python
# filter by min 5 innings
pp_df = pp_df[pp_df.innings >= 8]
pp_df.reset_index(inplace = True, drop = True)
```

```python
# filter by min 5 innings
mid_df = mid_df[mid_df.innings >= 8]
mid_df.reset_index(inplace = True, drop = True)
```

```python
# filter by min 5 innings
dth_df = dth_df[dth_df.innings >= 8]
dth_df.reset_index(inplace = True, drop = True)
```

### Calculation - Normalizing all values to same dimensions

```python
#step1: square of all values power play
pp_df['calc_SR'] = pp_df['SR'].apply(lambda x: x*x) 
pp_df['calc_RPI'] = pp_df['RPI'].apply(lambda x: x*x) 
pp_df['calc_BPD'] = pp_df['BPD'].apply(lambda x: x*x) 
pp_df['calc_BPB'] = pp_df['BPB'].apply(lambda x: x*x)

sq_sr, sq_rpi, sq_bpd, sq_bpb = np.sqrt(pp_df[['calc_SR','calc_RPI', 'calc_BPD', 'calc_BPB']].sum(axis = 0))

pp_df['calc_SR'] = pp_df['calc_SR'].apply(lambda x: x/sq_sr) 
pp_df['calc_RPI'] = pp_df['calc_RPI'].apply(lambda x: x/sq_rpi) 
pp_df['calc_BPD'] = pp_df['calc_BPD'].apply(lambda x: x/sq_bpd) 
pp_df['calc_BPB'] = pp_df['calc_BPB'].apply(lambda x: x/sq_bpb)

pp_df['calc_SR'] = pp_df['calc_SR'].apply(lambda x: x*wt_sr) 
pp_df['calc_RPI'] = pp_df['calc_RPI'].apply(lambda x: x*wt_rpi) 
pp_df['calc_BPD'] = pp_df['calc_BPD'].apply(lambda x: x*wt_bpd) 
pp_df['calc_BPB'] = pp_df['calc_BPB'].apply(lambda x: x*wt_bpb)

best_sr, worst_sr = max(pp_df['calc_SR']), min(pp_df['calc_SR'])
best_rpi, worst_rpi = max(pp_df['calc_RPI']), min(pp_df['calc_RPI'])
best_bpd, worst_bpd = max(pp_df['calc_BPD']), min(pp_df['calc_BPD'])
best_bpb, worst_bpb = min(pp_df['calc_BPB']), max(pp_df['calc_BPB'])
```

```python
# Step 1: Square of all values midle power
mid_df['calc_SR'] = mid_df['SR'].apply(lambda x: x*x) 
mid_df['calc_RPI'] = mid_df['RPI'].apply(lambda x: x*x) 
mid_df['calc_BPD'] = mid_df['BPD'].apply(lambda x: x*x) 
mid_df['calc_BPB'] = mid_df['BPB'].apply(lambda x: x*x)

# Step 2: Compute the square root of sum of squares
sq_sr, sq_rpi, sq_bpd, sq_bpb = np.sqrt(mid_df[['calc_SR', 'calc_RPI', 'calc_BPD', 'calc_BPB']].sum(axis=0))

# Step 3: Normalize by dividing by computed square roots
mid_df['calc_SR'] = mid_df['calc_SR'].apply(lambda x: x/sq_sr) 
mid_df['calc_RPI'] = mid_df['calc_RPI'].apply(lambda x: x/sq_rpi) 
mid_df['calc_BPD'] = mid_df['calc_BPD'].apply(lambda x: x/sq_bpd) 
mid_df['calc_BPB'] = mid_df['calc_BPB'].apply(lambda x: x/sq_bpb)

# Step 4: Multiply by respective weights
mid_df['calc_SR'] = mid_df['calc_SR'].apply(lambda x: x * wt_sr) 
mid_df['calc_RPI'] = mid_df['calc_RPI'].apply(lambda x: x * wt_rpi) 
mid_df['calc_BPD'] = mid_df['calc_BPD'].apply(lambda x: x * wt_bpd) 
mid_df['calc_BPB'] = mid_df['calc_BPB'].apply(lambda x: x * wt_bpb)

# Step 5: Determine best and worst values
best_sr, worst_sr = max(mid_df['calc_SR']), min(mid_df['calc_SR'])
best_rpi, worst_rpi = max(mid_df['calc_RPI']), min(mid_df['calc_RPI'])
best_bpd, worst_bpd = max(mid_df['calc_BPD']), min(mid_df['calc_BPD'])
best_bpb, worst_bpb = min(mid_df['calc_BPB']), max(mid_df['calc_BPB'])
```

```python
# Step 1: Square of all values death over
dth_df['calc_SR'] = dth_df['SR'].apply(lambda x: x*x) 
dth_df['calc_RPI'] = dth_df['RPI'].apply(lambda x: x*x) 
dth_df['calc_BPD'] = dth_df['BPD'].apply(lambda x: x*x) 
dth_df['calc_BPB'] = dth_df['BPB'].apply(lambda x: x*x)

# Step 2: Compute the square root of sum of squares
sq_sr, sq_rpi, sq_bpd, sq_bpb = np.sqrt(dth_df[['calc_SR', 'calc_RPI', 'calc_BPD', 'calc_BPB']].sum(axis=0))

# Step 3: Normalize by dividing by computed square roots
dth_df['calc_SR'] = dth_df['calc_SR'].apply(lambda x: x/sq_sr) 
dth_df['calc_RPI'] = dth_df['calc_RPI'].apply(lambda x: x/sq_rpi) 
dth_df['calc_BPD'] = dth_df['calc_BPD'].apply(lambda x: x/sq_bpd) 
dth_df['calc_BPB'] = dth_df['calc_BPB'].apply(lambda x: x/sq_bpb)

# Step 4: Multiply by respective weights
dth_df['calc_SR'] = dth_df['calc_SR'].apply(lambda x: x * wt_sr) 
dth_df['calc_RPI'] = dth_df['calc_RPI'].apply(lambda x: x * wt_rpi) 
dth_df['calc_BPD'] = dth_df['calc_BPD'].apply(lambda x: x * wt_bpd) 
dth_df['calc_BPB'] = dth_df['calc_BPB'].apply(lambda x: x * wt_bpb)

# Step 5: Determine best and worst values
best_sr, worst_sr = max(dth_df['calc_SR']), min(dth_df['calc_SR'])
best_rpi, worst_rpi = max(dth_df['calc_RPI']), min(dth_df['calc_RPI'])
best_bpd, worst_bpd = max(dth_df['calc_BPD']), min(dth_df['calc_BPD'])
best_bpb, worst_bpb = min(dth_df['calc_BPB']), max(dth_df['calc_BPB'])
```

###  Calculation - Comparision against the best & worst values

```python
pp_df['dev_best_SR'] = pp_df['calc_SR'].apply(lambda x: (x-best_sr)*(x-best_sr)) 
pp_df['dev_best_RPI'] = pp_df['calc_RPI'].apply(lambda x: (x-best_rpi)*(x-best_rpi)) 
pp_df['dev_best_BPD'] = pp_df['calc_BPD'].apply(lambda x: (x-best_bpd)*(x-best_bpd)) 
pp_df['dev_best_BPB'] = pp_df['calc_BPB'].apply(lambda x: (x-best_bpb)*(x-best_bpb))

pp_df['dev_best_sqrt'] = pp_df.apply(lambda x: x['dev_best_SR'] + x['dev_best_RPI'] + x['dev_best_BPD'] + x['dev_best_BPB'], axis = 1) 

pp_df['dev_worst_SR'] = pp_df['calc_SR'].apply(lambda x: (x-worst_sr)*(x-worst_sr)) 
pp_df['dev_worst_RPI'] = pp_df['calc_RPI'].apply(lambda x: (x-worst_rpi)*(x-worst_rpi)) 
pp_df['dev_worst_BPD'] = pp_df['calc_BPD'].apply(lambda x: (x-worst_bpd)*(x-worst_bpd)) 
pp_df['dev_worst_BPB'] = pp_df['calc_BPB'].apply(lambda x: (x-worst_bpb)*(x-worst_bpb))

pp_df['dev_worst_sqrt'] = pp_df.apply(lambda x: x['dev_worst_SR'] + x['dev_worst_RPI'] + x['dev_worst_BPD'] + x['dev_worst_BPB'], axis = 1) 
```

```python
mid_df['dev_best_SR'] = mid_df['calc_SR'].apply(lambda x: (x - best_sr) ** 2) 
mid_df['dev_best_RPI'] = mid_df['calc_RPI'].apply(lambda x: (x - best_rpi) ** 2) 
mid_df['dev_best_BPD'] = mid_df['calc_BPD'].apply(lambda x: (x - best_bpd) ** 2) 
mid_df['dev_best_BPB'] = mid_df['calc_BPB'].apply(lambda x: (x - best_bpb) ** 2)

mid_df['dev_best_sqrt'] = mid_df.apply(lambda x: x['dev_best_SR'] + x['dev_best_RPI'] + x['dev_best_BPD'] + x['dev_best_BPB'], axis=1) 

mid_df['dev_worst_SR'] = mid_df['calc_SR'].apply(lambda x: (x - worst_sr) ** 2) 
mid_df['dev_worst_RPI'] = mid_df['calc_RPI'].apply(lambda x: (x - worst_rpi) ** 2) 
mid_df['dev_worst_BPD'] = mid_df['calc_BPD'].apply(lambda x: (x - worst_bpd) ** 2) 
mid_df['dev_worst_BPB'] = mid_df['calc_BPB'].apply(lambda x: (x - worst_bpb) ** 2)

mid_df['dev_worst_sqrt'] = mid_df.apply(lambda x: x['dev_worst_SR'] + x['dev_worst_RPI'] + x['dev_worst_BPD'] + x['dev_worst_BPB'], axis=1)
```

```python
dth_df['dev_best_SR'] = dth_df['calc_SR'].apply(lambda x: (x - best_sr) ** 2) 
dth_df['dev_best_RPI'] = dth_df['calc_RPI'].apply(lambda x: (x - best_rpi) ** 2) 
dth_df['dev_best_BPD'] = dth_df['calc_BPD'].apply(lambda x: (x - best_bpd) ** 2) 
dth_df['dev_best_BPB'] = dth_df['calc_BPB'].apply(lambda x: (x - best_bpb) ** 2)

dth_df['dev_best_sqrt'] = dth_df.apply(lambda x: x['dev_best_SR'] + x['dev_best_RPI'] + x['dev_best_BPD'] + x['dev_best_BPB'], axis=1) 

dth_df['dev_worst_SR'] = dth_df['calc_SR'].apply(lambda x: (x - worst_sr) ** 2) 
dth_df['dev_worst_RPI'] = dth_df['calc_RPI'].apply(lambda x: (x - worst_rpi) ** 2) 
dth_df['dev_worst_BPD'] = dth_df['calc_BPD'].apply(lambda x: (x - worst_bpd) ** 2) 
dth_df['dev_worst_BPB'] = dth_df['calc_BPB'].apply(lambda x: (x - worst_bpb) ** 2)

dth_df['dev_worst_sqrt'] = dth_df.apply(lambda x: x['dev_worst_SR'] + x['dev_worst_RPI'] + x['dev_worst_BPD'] + x['dev_worst_BPB'], axis=1)
```

### Final Step - Overall score for each player ( values from 0 - 1)

```python
pp_df['score'] = pp_df.apply(lambda x: x['dev_worst_sqrt']/(x['dev_worst_sqrt'] + x['dev_best_sqrt']), axis = 1)
```

```python
mid_df['score'] = mid_df.apply(lambda x: x['dev_worst_sqrt']/(x['dev_worst_sqrt'] + x['dev_best_sqrt']), axis = 1)
```

```python
dth_df['score'] = dth_df.apply(lambda x: x['dev_worst_sqrt']/(x['dev_worst_sqrt'] + x['dev_best_sqrt']), axis = 1)
```
