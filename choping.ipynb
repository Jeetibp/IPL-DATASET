```python
import pandas as pd

# Load the CSV file
file_path = "deliveries.csv"
data = pd.read_csv(file_path)

# Exclude rows where match_id is within the specified range
filtered_ballbyball = data[(data['match_id'] < 335982) | (data['match_id'] > 1312200)]

# Save the remaining data to a new CSV file
filtered_ballbyball.to_csv("remaining_ballbyball.csv", index=False)

print("Rows with match_id in the range 335982 to 1312200 have been deleted. remaining data saved to 'remaining_ballbyball.csv'.")
```
